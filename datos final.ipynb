```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Empezamos con el ingreso

```python
ynr = pd.read_csv('datos finales\ingr_region_mes.csv')
```

```python
ynr['mean'] = ynr.mean(axis=1)
```

```python
ynr.columns
```

```python
ynr['log_ingr'] = np.log(ynr['mean'])
ynr['log_ingr20'] = np.log(ynr['percentile_20'])
ynr['log_ingr40'] = np.log(ynr['percentile_40'])
ynr['log_ingr60'] = np.log(ynr['percentile_60'])
ynr['log_ingr80'] = np.log(ynr['percentile_80'])
```

```python
percentiles = [20, 40, 60, 80]
percentile_values = ynr['mean'].quantile([p / 100 for p in percentiles])

for p, value in zip(percentiles, percentile_values):
    ynr[f'percentile_{p}'] = value
```

```python
ynr
```

```python
ynr[['log_ingr', 'log_ingr20', 'log_ingr40', 'log_ingr60', 'log_ingr80']].plot()
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Percentiles and log_ingr')
plt.legend(loc='upper left')
plt.show()
```
